```python
import pandas as pd
import os
```

```python
census_csv = os.path.join("Resources","sc-est2018-alldata6.csv")
census_raw = pd.read_csv(census_csv)
```

```python
census_raw.head()
```

```python
census_2017 = census_raw[["STATE", "NAME", "SEX", "ORIGIN", "RACE", "AGE","POPESTIMATE2017"]]
census_2017 = census_2017.rename(columns={"STATE":"STATE ID", "NAME":"STATE NAME", "POPESTIMATE2017": "2017 POPULATION"})
```

```python
census_2017.head()
```

```python
sex_deduped = census_2017.loc[census_2017["SEX"] == 0]
sex_deduped
```

```python
origin_deduped = sex_deduped.loc[census_2017["ORIGIN"] == 0]
origin_deduped
```

```python
tot_population = origin_deduped.groupby(["STATE NAME"])["2017 POPULATION"].sum()

state_population = pd.DataFrame({"2017 Population": tot_population})
state_population.reset_index()
```

```python
total_us_population = state_population.sum(axis = 0, skipna = True) 
total_us_population
```
