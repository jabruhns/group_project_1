```python
# Dependencies and Setup
import pandas as pd

# load file
file_to_load = "Resources/2017-PIT-Counts-by-State.csv"

# Read  File and store into Pandas data frame
homeless_data = pd.read_csv(file_to_load)

homeless_data.head()
```

```python
# get necessary columns
homeless_2017 = homeless_data[["State", "Overall Homeless, 2017"]]
homeless_2017 = homeless_2017.rename(columns={"State": "STATE ID", "Overall Homeless, 2017": "2017 Homeless Population"})
homeless_2017.dropna(inplace=True)
homeless_2017 = homeless_2017.set_index("STATE ID")

homeless_2017
```

```python
#Print to csv
homeless_2017.to_csv(r'Resources/homelessness_by_state.csv')
```

```python
#Dividing by Regions
region_1_states = (["CT", "ME", "MA", "NH",
                   "NJ", "NY", "PA", "RI", "VT"])

region1 = []

for a in region_1_states:
    region1.append(homeless_2017.loc[a, "2017 Homeless Population"])

Region1_homelessness = pd.DataFrame({
    "State ID" : region_1_states,
    "Homeless Population (2017)": region1
})




region_2_states = (["IL", "IN", "IA", "KS",
                   "MI", "MN", "MO", "NE", "ND", "OH", "SD"])

region2 = []

for a in region_2_states:
    region2.append(homeless_2017.loc[a, "2017 Homeless Population"])

Region2_homelessness = pd.DataFrame({
    "State ID" : region_2_states,
    "Homeless Population (2017)": region2
})

  

    
region_3_states = (["AL", "AR", "DE", "DC",
                   "FL", "GA", "KY", "LA", "MD", "MS", 
                    "NC", "OK", "SC", "TN", "TX"])

region3 = []

for a in region_3_states:
    region3.append(homeless_2017.loc[a, "2017 Homeless Population"])

Region3_homelessness = pd.DataFrame({
    "State ID" : region_3_states,
    "Homeless Population (2017)": region3
})
   


    
region_4_states = (["AK", "AZ", "CA", "HI",
                   "ID", "MT", "NV", "NM", "OR", "UT", 
                    "WA"])

region4 = []

for a in region_4_states:
    region4.append(homeless_2017.loc[a, "2017 Homeless Population"])

Region4_homelessness = pd.DataFrame({
    "State ID" : region_4_states,
    "Homeless Population (2017)": region4
})

```
