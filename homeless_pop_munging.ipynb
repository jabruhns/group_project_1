```python
# Dependencies and Setup
import pandas as pd

# load file
file_to_load = "Resources/2017-PIT-Counts-by-State.csv"

# Read  File and store into Pandas data frame
homeless_data = pd.read_csv(file_to_load)

homeless_data.head()
```

```python
# get necessary columns
homeless_2017 = homeless_data[["State", "Overall Homeless, 2017"]]
homeless_2017 = homeless_2017.rename(columns={"State": "STATE ID", "Overall Homeless, 2017": "2017 Homeless Population"})
homeless_2017.dropna(inplace=True)
homeless_2017 = homeless_2017.set_index("STATE ID")

homeless_2017_pd = pd.to_numeric(homeless_2017["2017 Homeless Population"].str.replace(',',''), errors='coerce')
homeless_2017_pd
```

```python
Region1_sum = homeless_2017_pd["CT"] + homeless_2017_pd["ME"] + homeless_2017_pd["MA"] + homeless_2017_pd["NH"] + homeless_2017_pd["NJ"] + homeless_2017_pd["NY"] + homeless_2017_pd["PA"] + homeless_2017_pd["RI"] + homeless_2017_pd["VT"]
Region2_sum = homeless_2017_pd["IL"] + homeless_2017_pd["IN"] + homeless_2017_pd["IA"] + homeless_2017_pd["KS"] + homeless_2017_pd["MI"] + homeless_2017_pd["MN"] + homeless_2017_pd["MO"] + homeless_2017_pd["NE"] + homeless_2017_pd["ND"] + homeless_2017_pd["OH"] + homeless_2017_pd["SD"]
Region3_sum = homeless_2017_pd["AL"] + homeless_2017_pd["AR"] + homeless_2017_pd["DE"] + homeless_2017_pd["DC"] + homeless_2017_pd["FL"] + homeless_2017_pd["GA"] + homeless_2017_pd["KY"] + homeless_2017_pd["LA"] + homeless_2017_pd["MD"] + homeless_2017_pd["MS"] + homeless_2017_pd["NC"] + homeless_2017_pd["OK"] + homeless_2017_pd["SC"] + homeless_2017_pd["TN"] + homeless_2017_pd["TX"]
Region4_sum = homeless_2017_pd["AK"] + homeless_2017_pd["AZ"] + homeless_2017_pd["CA"] + homeless_2017_pd["HI"] + homeless_2017_pd["ID"] + homeless_2017_pd["MT"] + homeless_2017_pd["NV"] + homeless_2017_pd["NM"] + homeless_2017_pd["OR"] + homeless_2017_pd["UT"] + homeless_2017_pd["WA"]

```

```python
homelessness_by_region = pd.DataFrame({
    "Region" : ["1", "2", "3", "4"],
    "Homeless Population (2017)": [Region1_sum, Region2_sum, Region3_sum, Region4_sum]
})

homelessness_by_region
```

```python
#Print to csv
homeless_2017.to_csv(r'Resources/homelessness_by_state.csv')
homelessness_by_region.to_csv(r'Resources/homelessness_by_region.csv')
```
